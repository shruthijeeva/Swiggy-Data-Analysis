## 2. Box Plot:

```python
sns.boxplot(data=df, y="City", x="Avg ratings")
plt.title("Box Plot")
plt.show()
```

## 3. Heatmap:

```python
corr = df[[
    "Price",
    "Avg ratings",
    "Total ratings",
    "Restaurant_name_length"
]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("HeatMap")
plt.show()
```

## 4. Histogram:

```python
sns.histplot(data=df, x="Price", kde=True, color="teal")
plt.title("Histogram")
plt.show()
```
